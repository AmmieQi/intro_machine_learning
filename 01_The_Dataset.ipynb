The dataset is INVENTED. I took the famous Wines dataset from http://archive.ics.uci.edu/ml/datasets/Wine, got rid of most of the features and changed others and invented one.

The dataset is now called ``fruit``. It allows us to compare apples to oranges! (Also apples to pears, since that's the French expression).

The columns are:

* ``fruit``: 1-3, numeric id for:
* ``fruit_name``: orange, pear and apple, corresponding to ``fruit`` == 1, 2, and 3, respectively
* ``color``: 1-6, numeric id for:
* ``color_name``: blue, brown, green, orange, red, yellow, corresponding to ``color`` == 1-6, respectively.
* ``compactness``: 0-1, continuous. A concept borrowed from the famous ``seeds`` dataset. If one were to take a two-dimensional image of the fruit (presumably in a random orientation) and make the smallest ellipse it would fit within, ``compactness`` is the length of the short axis divided by the length of the long axis. An infinitely long line has a compactness of 0, a perfect circle (or square, for that matter) has a compactness of 1.
* ``weight``: in grams
* ``sweetness``: in totally fictional units; I just took the (unit-unspecified) values for proline from ``Wines`` and fudged them a bit.
* ``acidity``: same note as ``sweetness``, except original column was OD280/315.

#### 3. A quick random forest classifier to see how 'good' the features are and their relative importane #

This is skipping ahead a bit.

```python
# code adapted from http://blog.yhathq.com/posts/random-forests-in-python.html

df.sort('fruit', inplace=True) # This is important because the factorizer assigns numbers
    # based on the order the first label is encountered, e.g. if the first instance had
    # fruit = 3, the y value would be 0.
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75 # randomly assign training and testing set
train, test = df[df['is_train']==True], df[df['is_train']==False]
features = ['color', 'compactness', 'weight', 'sweetness', 'acidity']
clf = RandomForestClassifier(n_jobs=2)
y, _ = pd.factorize(train['fruit'])
clf.fit(train[features], y)
preds = clf.predict(test[features])
pd.crosstab(test['fruit'], preds, rownames=['actual'], colnames=['preds'])
```

The above confusion matrix should show close to or equal to 100% accurate classification.

Here are the feature importances (%):

```python
for i, score in enumerate(list(clf.feature_importances_)):
    print(round(100*score, 1), features[i])
```
