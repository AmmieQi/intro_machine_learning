```python
# Create numpy arrays for classifier
X = []
y = []
for i, row in df.iterrows():
    X.append([row.sweetness_normal, row.acidity_normal])
    y.append(row.fruit)
X = np.array(X)
y = np.array(y)

for n_neighbors in [1, 3, 5, 7, 9, 11, 99, 131, 177]:

    # import some data to play with

    h = 0.02  # step size in the mesh

    # Create color maps
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors)

    clf.fit(X, y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("K-nearest neighbors of entire dataset (k = %i)"
              % (n_neighbors))

    plt.show()
```

# Showing training and testing sets

```python
# Divide 70:30 train:test

eval_x = [1, 7, 13, 13, 13, 31, 71]

for k in eval_x:
    df['is_train'] = np.random.uniform(0, 1, len(df)) <= .70
    train, test = df[df['is_train']==True], df[df['is_train']==False]

    X = []
    y = []
    for i, row in train.iterrows():
        X.append([row.sweetness_normal, row.acidity_normal])
        y.append(row.fruit)
    X = np.array(X)
    y = np.array(y)

    # import some data to play with

    h = 0.1  # step size in the mesh

    # Create color maps
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(k)

    clf.fit(X, y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    #evaluate the training points
    trainpred = clf.predict(train[['sweetness_normal', 'acidity_normal']])
    train['pred'] = [x for x in trainpred]
    train['correct'] = 0
    for idx, row in train.iterrows():
        if row.fruit == row.pred:
            train.loc[idx, 'correct'] = True
        else:
            train.loc[idx, 'correct'] = False

    # Predict on the testing points
    preds = clf.predict(test[['sweetness_normal', 'acidity_normal']])
    test['pred'] = [x for x in preds]
    test['correct'] = 0
    for idx, row in test.iterrows():
        if row.fruit == row.pred:
            test.loc[idx, 'correct'] = True
        else:
            test.loc[idx, 'correct'] = False

    #plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, alpha=0.5)
    #plt.scatter(train[train.correct==True].sweetness_normal, train[train.correct==True].acidity_normal, c=train[train.correct==True].fruit, cmap=cmap_bold, alpha=0.2)
    #plt.scatter(train[train.correct==False].sweetness_normal, train[train.correct==False].acidity_normal, c=train[train.correct==False].fruit, cmap=cmap_bold, alpha=1)
    plt.scatter(test[test.correct==True].sweetness_normal, test[test.correct==True].acidity_normal, s=20, c=test[test.correct==True].pred, cmap=cmap_bold, alpha=0.2)
    plt.scatter(test[test.correct==False].sweetness_normal, test[test.correct==False].acidity_normal, s=60, c=test[test.correct==False].fruit, cmap=cmap_bold, alpha=1)
    #, c=test.fruit)#, cmap=cmap_bold, alpha=1)
    # Plot testing points

    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("K-NN showing test set (k = %i)"
              % (k))

    plt.show()
```

```python
eval_x = [1,3,5,7,9,11,15,19,23,27,31,37,41,47,51,57,61,71,81,91,111,131,151]
y_train = []
y_test = []

for k in eval_x:
    
    print("k={}".format(k), end=" ")

    sumtrain = 0
    sumtest = 0
    
    for rep in range(10):
        
        print(rep, end=" ")

        df['is_train'] = np.random.uniform(0, 1, len(df)) <= .70
        train, test = df[df['is_train']==True], df[df['is_train']==False]

        X = []
        y = []
        for i, row in train.iterrows():
            X.append([row.sweetness_normal, row.acidity_normal])
            y.append(row.fruit)
        X = np.array(X)
        y = np.array(y)

        # import some data to play with

        h = 0.1  # step size in the mesh

        # Create color maps
        cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
        cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

        # we create an instance of Neighbours Classifier and fit the data.
        clf = neighbors.KNeighborsClassifier(k)

        clf.fit(X, y)

        #evaluate the training points
        trainpred = clf.predict(train[['sweetness_normal', 'acidity_normal']])
        train['pred'] = [x for x in trainpred]
        train['correct'] = 0
        for idx, row in train.iterrows():
            if row.fruit == row.pred:
                train.loc[idx, 'correct'] = True
            else:
                train.loc[idx, 'correct'] = False

        # Predict on the testing points
        preds = clf.predict(test[['sweetness_normal', 'acidity_normal']])
        test['pred'] = [x for x in preds]
        test['correct'] = 0
        for idx, row in test.iterrows():
            if row.fruit == row.pred:
                test.loc[idx, 'correct'] = True
            else:
                test.loc[idx, 'correct'] = False
    
        sumtrain += len(train[train.correct==False]) / (len(train[train.correct==True])+len(train[train.correct==False]))
        sumtest += len(test[test.correct==False]) / (len(test[test.correct==True])+len(test[test.correct==False]))
    
    y_train.append(sumtrain/10)
    y_test.append(sumtest/10)
                                                      
```

```python
plt.plot(eval_x, y_train, 'b-')
plt.plot(eval_x, y_test, 'r-')
```
